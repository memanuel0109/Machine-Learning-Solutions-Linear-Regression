```python
import pandas as pd
import folium

df = pd.read_csv('../Datasets/combined_cluster.csv', index_col=0)

chicago_map = folium.Map(location=[41.8781,-87.6298], tiles="CartoDB dark_matter")

colors = ['red', 'orange', 'darkgreen', 'lightgreen',
          'green', 'blue', 'darkblue', 'purple',
          'darkred', 'lightred', 'beige', 'lightgray',
          'pink']

for x, y, price, value in df.values:
    for index, color in enumerate(colors):
        if value == index:
            c = colors[index]
        
    folium.Marker(location=[x,y],
                  icon=folium.Icon(color=c,icon_color='white', icon='home'),
                  popup=price).add_to(chicago_map)

display(chicago_map)
```
